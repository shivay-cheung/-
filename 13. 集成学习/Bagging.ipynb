```python
import numpy  as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import neighbors #knn
from sklearn.ensemble import BaggingClassifier #bagging分类
from sklearn import tree #决策树
from sklearn.model_selection import train_test_split

#加载数据
iris = datasets.load_iris()
x_data = iris.data[:,:2]
y_data = iris.target

#分割数据
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data)

def plot(model):
    #获取数据值所在的范围
    x_min,x_max = x_data[:,0].min()-1,x_data[:,0].max()+1
    y_min,y_max = x_data[:,1].min()-1,x_data[:,1].max()+1
    
    #生成网格矩阵
    xx,yy = np.meshgrid(np.arange(x_min,x_max,0.02),
                       np.arange(y_min,y_max,0.02))
    z = model.predict(np.c_[xx.ravel(),yy.ravel()])
    z = z.reshape(xx.shape)
    #画出等高线
    cs = plt.contourf(xx,yy,z)

##输入数据 ， 创建模型
## -----------  KNN 算法 -----------
knn = neighbors.KNeighborsClassifier()
knn.fit(x_train,y_train)
plot(knn)

#样本散点图
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()

#准确率
print("KNN的准确率：",knn.score(x_test,y_test))

## -----------  决策树 算法 -----------
dtree = tree.DecisionTreeClassifier()
dtree.fit(x_train,y_train)
plot(dtree)

#样本散点图
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()

#准确率
print("决策树的准确率：",dtree.score(x_test,y_test))

## -----------  集成学习_KNN -----------
bagging_knn = BaggingClassifier(knn,n_estimators=100)
bagging_knn.fit(x_train,y_train)
plot(bagging_knn)

#样本散点图
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()

#准确率
print("集成学习_KNN的准确率：",bagging_knn.score(x_test,y_test))

## -----------  集成学习_决策树-----------
bagging_tree = BaggingClassifier(dtree)
bagging_tree.fit(x_train,y_train)
plot(bagging_tree)

#样本散点图
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()

#准确率
print("集成学习_决策树的准确率：",bagging_tree.score(x_test,y_test))

```
